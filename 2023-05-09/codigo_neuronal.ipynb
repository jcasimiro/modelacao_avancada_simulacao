```python

import numpy as np
import math
import matplotlib.pyplot as plt
from openpyxl import load_workbook
import pandas as pd
from scipy.stats import logistic
#%matplotlib inline
plt.rcParams['figure.figsize'] = [9, 6]

```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(X):
    return np.maximum(0,X)

def relud(X):
    return np.maximum(0,np.sign(X))

```

```python
#   DADOS TREINO (EXEMPLO)
X = [[23,12, 2,31,22,16,26,16,30,32], # temperatura e vento
    [ 5, 20, 2,31, 2, 4, 4,20,20,30]]
Y = [[ 1, 0, 0, 0, 1, 0, 1, 0, 1, 0]] # ir à praia
X=np.array(X)
print(np.shape(X))
Y=np.array(Y)
np.shape(Y)


# DADOS TREINO



```

```python
##                   UM NEURONIO 


#   COEFICIENTES NEURONIO
W=[0.5,0.5]             #2 FEATURES, 1 NEURONIO
W=np.array(W)
W=np.reshape(W,(2,1))
np.shape(W)
b=1

```

```python
for i in range(10000):
    #   ESTIMADOR LINEAR
    Z=np.dot(W.T,X)+b

    #   CLASSIFICAÇÃO BINÁRIA (PROBABLIDADE)
    A=logistic.cdf(Z)


    #   ERROS (#diferenças para o real)
    dZ = A - Y

    # 10 elementos no conjunto de treino, derivadas nos coeficientes para otimizar
    dW=1/10*np.dot(X,dZ.T)  ## 10 = len(X[1])

    db=1/10*np.sum(dZ)

    lr=0.002 #learning rate

    #   COEFICIENTES ATUALIZADOS
    W=W-lr*dW
    b=b-lr*db

print('Coef. Rede de 1 Neuronio', (W,b))
print('Erros', dZ)
```

```python

print('-')
print('3 Neuronios')
#           3 NEURONIOS

def tres_neuronios(X, Y,
                   W1, b1, #COEF NEURONIOS
                   W2, b2,
                   iteracoes=10000,
                   max_error=False,
                   erro_maximo = 10**-6,
                   var_uso = False,
                   dados_uso = []):

    erro = False

    for i in range(iteracoes):

        Z1=np.dot(W1,X)+b1
        
        A1=relu(Z1)

        Z2=np.dot(W2,A1)+b2

        A2=logistic.cdf(Z2)

        dZ2=A2-Y

        dW2=1/10*np.dot(dZ2,A1.T)

        np.shape(dW2)
        db2=1/10*np.sum(dZ2)

        dZ1=np.multiply(np.dot(W2.T,dZ2),(relud(Z1)))

        np.shape(dZ1)
        dW1=1/10*np.dot(dZ1,X.T)

        np.shape(dW1)
        db1=1/10*np.sum(dZ1)

        lr=0.003 #learning rate
        W1=W1-lr*dW1
        b1=b1-lr*db1
        W2=W2-lr*dW2
        b2=b2-lr*db2

        ## Media erros
        media_erros = np.mean(dZ2)

        if max_error == True:
            if abs(media_erros) < erro_maximo:
                erro = True
                break

    if erro == True and max_error == True:
        print('Obtemos, em média, erros menor a ', erro_maximo)
    elif erro == False and max_error == True:
        print('Não foram iterações suf. para ter erro menor ao maximo')

    print(W1, b1)
    print(W2, b2)
    print('Errors', dZ2)


    #USO
    if var_uso == True:
        Z=np.dot(W.T,dados_uso)+b
        A=logistic.cdf(Z)

        return print('Prob.', A)

    return dZ2
```
